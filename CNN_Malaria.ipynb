```python
from tensorflow.keras.layers import Input, Conv2D,MaxPooling2D, Dropout,BatchNormalization, Dense,Flatten
```

```python
from tensorflow.keras.models import Model
```

```python
feed = Input(shape= (100,100,3) )
```

```python
cv1 = Conv2D(32, (3,3), activation ="relu", strides= (1,1), padding = "same") (feed)
```

```python
max1 = MaxPooling2D(pool_size=(2,2), padding ="valid")(cv1)
```

```python
cv2 = Conv2D(32, (3,3), activation ="relu", strides= (1,1), padding = "same") (max1)
```

```python
max2 = MaxPooling2D(pool_size=(2,2), padding ="valid")(cv2)
```

```python
cv3 = Conv2D(32, (3,3), activation ="relu", strides= (1,1), padding = "same") (max2)
```

```python
max3 = MaxPooling2D(pool_size=(2,2), padding ="valid")(cv3)
```

```python
flat = Flatten()(max3)
```

```python
den1 = Dense(256, activation = "relu")(flat)
```

```python
den2 = Dense(128, activation = "relu")(den1)
```

```python
den3 = Dense(256, activation = "relu")(den2)
```

```python
out = Dense(2, activation = "sigmoid") (den3)
```

```python
fmodel = Model(inputs = feed , outputs = out)
```

```python
fmodel.summary()
```

```python
fmodel.compile(optimizer = "adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"] )
```

```python
fmodel.fit(X_train,y_train, epochs = 20)
```

```python
fmodel.evaluate(X_test,y_test)
```
